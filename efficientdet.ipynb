```python
detector = hub.load("https://kaggle.com/models/tensorflow/efficientdet/frameworks/tensorflow2/variations/lite0-feature-vector/versions/1")
```

```python
layer = hub.KerasLayer(detector, trainable=True)
```

```python
type(layer)
```

```python
model = tf.keras.Sequential([
    layer
])
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.001,
    global_clipnorm=10.0,
)

model.compile(
    optimizer=optimizer, loss=SparseCategoricalCrossentropy()
)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite = converter.convert()
```

```python
def process_image(image_path, img_size=224):
  # image = tf.io.read_file(image_path)
  # image = tf.image.decode_image(image, channels=3)
  # image = tf.image.resize(image, (320, 320))
  # image = tf.cast(image, tf.float32)
  # image = tf.expand_dims(image, axis=0)
  # # Convert the colour channel values from 0-255 to 0-1 values
  # #image = tf.image.convert_image_dtype(image, tf.float32)[tf.newaxis, ...]
  # #image = tf.image.resize(image, size=[img_size, img_size])

  image = tf.io.read_file(image_path)

# Decode the image to a tensor with 3 channels (RGB)
  image = tf.image.decode_image(image, channels=3)

# Resize the image to the required shape (320x320)
  image = tf.image.resize(image, (320, 320))

# Normalize the pixel values to be in the range [0, 1]
  image = tf.cast(image, tf.float32) / 255.0

# Add batch dimension (None represents the batch size, which can be varied)
  image = tf.expand_dims(image, axis=0)

  return image
```

```python
image_path = './only_cars/test/images/frame_0441.jpg'
image_tensor = process_image(image_path, 512)
```
